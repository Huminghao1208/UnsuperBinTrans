This file load the embedding files generated by word2vec

```python
import numpy as np
```

```python
#data_path="data/processed/de-en/train.de.pth"
embedding_file = 'data/x86-vector.bin'
```

```python
def load_embeddings(file_path):
    with open(file_path, 'rb') as file:
        header = file.readline().decode('utf-8')
        vocab_size, vector_size = map(int, header.split())
        
        embeddings = {}
        binary_len = np.dtype('float32').itemsize * vector_size
        
        for _ in range(vocab_size):
            word = b''
            while True:
                ch = file.read(1)
                if ch == b' ':
                    break
                if ch != b'\n':
                    word += ch
            word = word.decode('utf-8')
            
            vector = np.fromstring(file.read(binary_len), dtype='float32')
            embeddings[word] = vector
    
    return embeddings
```

```python
def main():
    word_embeddings = load_embeddings(embedding_file)

# Print the loaded word embeddings
    for word, vector in word_embeddings.items():
        print(f"Word: {word}")
        print(f"vector length is: {len(vector)}")
        #print(f"Embedding: {vector}")
        #print()
if __name__ == '__main__':
    main()        
```
